## Задание 1
Даны два списка, в списках допустимы повторяющимися элементы. Определить входит ли  
первый во второй список с учётом количества повторяющихся элементов. Без использования  
словарей.  
Примеры:  
[1, 2, 3, 3, 4] [1, 2, 3, 7, 4, 3] – входит  
[1, 2, 3, 3, 4, 3] [1, 2, 3, 7, 4, 3] – не входит  
[1, 2, 3, 3, 8, 4] [1, 2, 3, 7, 4, 3] – не входит  
[1, 2, 3, “3”, 4] [1, 2, 3, 7, 4, 3] – не входит

```python
def is_included(list1: list, list2: list):
    """Функция для проверки вхождения первого списка во второй.

    Args:
        list1 (list): первый список
        list2 (list): второй список

    Returns:
        str: "входит", или "не входит"
    """
    # оздаём копию второго списка
    list2_temp = list2[:]
    
    # Ищем каждый элемент первого списка в копии второго
    for x in list1:
        if x in list2_temp:
            # Удаляем найденный элемент (для учёта повторяющихся элементов)
            list2_temp.remove(x)
        else:
            return 'не входит'
        
    # Если хоть один не нашли, ответ "не входит", иначе - "входит".
    return 'входит'


# Тестовые данные
list1 = [1, 2, 3, 3, 4]
list2 = [1, 2, 3, 3, 4, 3]
list3 = [1, 2, 3, 3, 8, 4]
list4 = [1, 2, 3, '3', 4]
list5 = [1, 2, 3, 7, 4, 3]

# Сравниваем все списки с пятым (как в примерах к заданию)
print(list1, '\t', list5, ' – ', is_included(list1, list5))
print(list2, '\t', list5, ' – ', is_included(list2, list5))
print(list3, '\t', list5, ' – ', is_included(list3, list5))
print(list4, '\t', list5, ' – ', is_included(list4, list5))
```

## Задание 2
Даны два списка, в списках допустимы повторяющимися элементы, необходимо поменять их  
значения между собой без использования других списков и картежей.


```python
def swap_lists_items(list1: list, list2: list):
    """Функция для обмена значений двух списков между собой.

    Args:
        list1 (list): первый список
        list2 (list): второй список
    """
    # Сохраняем длины обоих списков
    list1_len = len(list1)
    list2_len = len(list2)
    
    # Сравниваем длины списков, в конец списка меньшей длины добавляем None, пока длины не сравняются
    if list1_len > list2_len:
        for _ in range(list1_len - list2_len):
            list2.append(None)
    elif list1_len < list2_len:
        for _ in range(list2_len - list1_len):
            list1.append(None)        
    
    # Меняем значения списков между собой
    for i in range(len(list1)):
        temp = list1[i]
        list1[i] = list2[i]
        list2[i] = temp
    
    # Удаляем добавленные None
    del list1[list2_len:]
    del list2[list1_len:]


# Тестовые данные
list1 = [1, 2, 2, 3, 4, 5, 5, 5, 6]
list2 = [2, 5, 9, 5, 9]

# Выводим списки и их адреса в памяти
print('list1\t', list1, '\t', id(list1))
print('list2\t', list2, '\t', id(list2))

# Выполняем функцию и выводим списки и их адреса в памяти ещё раз
swap_lists_items(list1, list2)
print('list1\t', list1, '\t', id(list1))
print('list2\t', list2, '\t', id(list2))
```

## Задание 3
Создать программу, которая умеет перемножать матрицы. Можно ограничиться матрицами  
не более чем 3x3.  
Матрицу 3x3 вполне можно записать так: matrix = ((1, 2, 3), (4, 5, 6), (7, 8, 9))


```python
def multiply_matrs(matr1, matr2):
    """Функция для перемножения двух матриц.

    Args:
        matr1 (list): первая матрица с количеством строк i и количеством столбцов k
        matr2 (list): вторая матрица с количеством строк k и количеством столбцов j

    Raises:
        ValueError: возвращает ошибку, если количество столбцов первой матрицы не равно количеству строк второй матрицы

    Returns:
        list: возвращает матрицу - произведение матриц matr1 и matr2 с количеством строк i и количеством столбцов j
    """
    # Проверяем, совпадает ли количество столбцов первой матрицы с количеством строк второй
    if len(matr1[0]) != len(matr2):
        raise ValueError('Количество столбцов первой матрицы должно быть равно числу строк второй матрицы!')
    
    # Сохраняем i, k, j
    i = len(matr1)
    k = len(matr2)
    j = len(matr2[0])

    # Создаём матрицу - произведение с количеством строк i и количеством столбцов j
    matr_result = [[0 for _ in range(j)] for _ in range(i)]

    # Перемножаем матрицы по итеративному алгоритму и записываем результат
    for i_index in range(i):
        for j_index in range(j):
            for k_index in range(k):
                matr_result[i_index][j_index] += matr1[i_index][k_index] * matr2[k_index][j_index]
        
    return matr_result


# Тестовые данные
matr1 = [[7,8,2],
        [0,8,9],
        [7,66,1],
        [4,5,3],
        [2,3,44]]
matr2 = [[1, 2],
        [4, 5],
        [7, 8]]

# Перемножаем матрицы и красиво выводим
matr_result = multiply_matrs(matr1, matr2)
for i in matr_result:
    print(i)
```
